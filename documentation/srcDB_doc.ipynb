# Source datasets

This Jupyter Notebook analyses the source datasets and the process of cleaning up for ["Blessed Be the Fruit"](https://github.com/OrsolaMBorrini/blessedfruit), an open data project regarding the analysis of factors that might influence pregnancy rates in young women in Italy.

## Selection phase
Overall, **7 source datasets** have been selected, all coming from [Istat](https://www.istat.it/) various databases and platforms such as [demo - demography in figures](https://demo.istat.it/?l=en), [IstatData](https://esploradati.istat.it/databrowser/#/) and [I.Stat](http://dati.istat.it/?lang=en).

When possible, the IstatData has been preferred to I.Stat, as all the content of the latter will be gradually migrated in the former.

## D3 - "Aspects of daily life: Religious observance - detail age"

### Analysis of the dataset D3

**Path** to the dataset on the I.stat platform: `Daily life and citizen opinions > Social activities and religious observances > Religious observances - detail age`

Select variables:
* Measure: thousands value
* Gender: females
* Age class: 14-17, 18-19, 20-24
* Select time: 2017, 2018, 2019
* Data type: "at least once a week", "never"


**Export in CSV** and get this table (some rows are shown as example):
<br>

TIPO_DATO_AVQ | Data type | MISURA_AVQ | Measure | SEXISTAT1 | Gender | ETA1 | Age class | TIME | Select time | Value | Flag Codes | Flags
--- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | ---|
6_NEVER_RELIG | never | THV | thousands value | 2 | females | Y14-17 | 14-17 years | 2017 | 2017 | 261 | |
6_WEEK_RELIG | at least once a week | THV | thousands value | 2 | females | Y18-19 | 18-19 years | 2018 | 2018 | 200 | |
6_WEEK_RELIG | at least once a week | THV | thousands value | 2 | females | Y20-24 | 20-24 years | 2019 | 2019 | 400 | |

<br>

We can see that we have repeated data:
* `TIPO_DATO_AVQ` corresponds to `Data type`
* `MISURA_AVQ` corresponds to `Measure`
* `SEXISTAT1` corresponds to `Gender`
* `ETA1` corresponds to `Age class`
* `TIME` corresponds to `Select time`

Therefore, we can get rid of these duplicated columns in order to keep our dataset cleaner.

### Disclaimer
We have decided to keep the analysis of the **Religious Observance** factor only on the national level because of two reasons:
1.   Lack of clear and usable data regarding religious observance and its variations (gender, age class, regions) on Istat
2.   Preference for the age class aspect with respect to the geographic aspect

The dataset "Aspects of daily life: Religious observances - regions and type of municipality" is therefore useless and will not be employed in this project.
